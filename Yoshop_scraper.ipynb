## Importing libraries

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

## Main soup 

```python
url = 'https://yoshops.com'
html = urlopen(url)
page_soup = BeautifulSoup(html,'html.parser')
print(page_soup.prettify())
```

### Function for categories and sub categories and their links

```python
def extractSubcategoryLinks(categorySoup):
  categorySubCategoryLinks={}
  for i in categorySoup:
    subCategories={}
    for j in i.find('ul').find_all('a'):
      subCategories[j.text]=j['href']
    categorySubCategoryLinks[i.find('a').text]=subCategories
  return categorySubCategoryLinks
```

### Extract Categories and sub-categories links

```python
categorySoup = page_soup.find("ul", {"class":"nav category-bar"}).find_all('li',{"class":"dropdown"})
categorySubCategoryLinks=extractSubcategoryLinks(categorySoup)
categorySubCategoryLinks
```

### Function for Product count

```python
def productCounter(categorySubCategoryLinks): 
  categorySubCategoryCount={}
  for i in categorySubCategoryLinks:
    subCategoryCount={}
    for j in categorySubCategoryLinks[i]:
      productSoup=BeautifulSoup(urlopen(url+categorySubCategoryLinks[i][j]),'html.parser').find_all('span',{"class":"product-title"})
      subCategoryCount[j]=len(productSoup)
    categorySubCategoryCount[i]=subCategoryCount
  return categorySubCategoryCount
```

### Extract product counts per sub category

```python
productCounter(categorySubCategoryLinks)
```
